```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix
from country_converter import CountryConverter
```

```python
rn_con_col_names = ['Country','Code','Year','Other','Solar','Wind','Hydro']
rn_cap_col_names = ['Country','Code','Year','Wind','Hydro','Solar','Other']
```

```python
rn_con = pd.read_csv('modern-renewable-energy-consumption.csv',
                     header=0,
                     names=rn_con_col_names)
rn_cap = pd.read_csv('modern-renewable-prod.csv',
                     header=0,
                     names=rn_cap_col_names)
```

```python
cc = CountryConverter()
```

```python
rn_con['Continent'] = rn_con['Code'].fillna('').astype(str).apply(lambda x: cc.convert(names=x, to='continent'))

rn_cap['Continent'] = rn_cap['Code'].fillna('').astype(str).apply(lambda x: cc.convert(names=x, to='continent'))
```

```python
rn_cap.head()
```

```python
rn_con_asia = rn_con[rn_con['Continent'] == 'Asia']
rn_cap_asia = rn_cap[rn_cap['Continent'] == 'Asia']
```

```python
print(rn_con_asia.describe())
print(rn_cap_asia.describe())
```

```python
# Filter data for years starting from 2010
start_year = 2010
rn_con_asia_2010 = rn_con_asia[rn_con_asia['Year'] >= start_year]

# Create a line plot for Hydro energy consumption over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Hydro', data=rn_con_asia_2010, hue='Country')
plt.title(f'Hydro Energy Consumption Over Time (Starting from {start_year})')
plt.xlabel('Year')
plt.ylabel('Hydro Energy Consumption')
plt.show()
```

```python
# Calculate total energy consumption for each country
rn_con_asia['Total_Consumption'] = rn_con_asia[['Hydro', 'Wind', 'Solar', 'Other']].sum(axis=1)
```

```python
# Calculate total energy production for each country
rn_cap_asia['Total_Production'] = rn_cap_asia[['Hydro', 'Wind', 'Solar', 'Other']].sum(axis=1)
```

```python
# Find the top 7 countries based on total energy consumption
top_countries = rn_con_asia.groupby('Country')['Total_Consumption'].sum().sort_values(ascending=False).head(7).index

# Filter the DataFrame for the top countries
rn_con_asia_7 = rn_con_asia[rn_con_asia['Country'].isin(top_countries)]

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Total_Consumption', data=rn_con_asia_7, estimator=sum, order=top_countries)
plt.title('Total Renewable Energy Consumption by Top 7 Countries in Asia')
plt.xlabel('Country')
plt.ylabel('Total Energy Consumption')
plt.show()
```

```python
scatter_cols = ['Hydro', 'Wind', 'Solar', 'Other']

scatter_matrix(rn_con_asia[scatter_cols], alpha=0.8, figsize=(12, 12), diagonal='hist')

plt.suptitle('Scatter Matrix of Energy Sources', y=0.95, fontsize=16)
plt.show()
```

```python
# Merge the consumption and production datasets
rn_asia_combined = pd.merge(rn_con_asia, rn_cap_asia, on=['Country', 'Year'])
```

```python
# Create a scatter plot to visualize the correlation
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Total_Consumption', y='Total_Production', data=rn_asia_combined, hue='Country')
plt.title('Correlation between Energy Consumption and Production')
plt.xlabel('Total Energy Consumption')
plt.ylabel('Total Energy Production')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
